```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df, min_support=0.6)
```

```python
fpgrowth(df, min_support=0.6, use_colnames=True)
```

FP Growth dirancang untuk beroperasi pada basis data yang berisi transaksi, seperti pembelian oleh pelanggan suatu toko. Suatu itemset dianggap "frequent" jika memenuhi ambang batas dukungan yang ditentukan pengguna. Misalnya, jika ambang batas dukungan ditetapkan bernilai 0,5 (50%), frequent itemset didefinisikan sebagai serangkaian item yang muncul bersamaan dalam setidaknya 50% dari semua transaksi dalam basis data.
